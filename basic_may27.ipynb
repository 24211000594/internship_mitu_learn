```python
# check id the number is in between 50 and 100 or not (use  nested if)
```

```python
num = int(input("Enter a number: "))

if num >= 50:
    if num <= 100:
        print("Number is between 50 and 100")
    else:
        print("Number is greater than 100")
else:
    print("Number is less than 50")
```

```python
# Check if number is between 50 and 100 using logical operator

num = int(input("Enter a number: "))

if num >= 50 and num <= 100:
    print("Number is between 50 and 100")
elif num < 50:
    print("Number is less than 50")
else:
    print("Number is greater than 100")
```

```python
# read purchase cost and selling cost from the user
# state profit or loss and how much
```

```python

purchase = float(input("Enter purchase cost: "))
selling = float(input("Enter selling cost: "))

if selling > purchase:
    profit = selling - purchase
    print("Profit of:", profit)
elif selling < purchase:
    loss = purchase - selling
    print("Loss of:", loss)
else:
    print("No profit, no loss")
```

```python
# Check if years are leap or not using nested if
```

```python

year = int(input("Enter the year:"))

if year % 4 == 0:
    if year % 100 == 0:
        if year % 400 == 0:
            print("Century Leap Year")
        else:
            print("Century but not a Leap Year")
    else:
        print("Leap Year")
else:
    print("NOT a Leap Year")
```

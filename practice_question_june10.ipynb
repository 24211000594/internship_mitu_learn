```python
## 2....

numbers = [12, 7, 23, 48, 15, 66, 9]
total = 0
for num in numbers:
    if num % 2 == 0:
        total += num

print("Sum of even numbers:", total)
```

```python
## 3....


data = ['apple', 'banana', 'cherry', 'date', 'elderberry']
reversed_list = []

for i in range(len(data) - 1, -1, -1):
    reversed_list.append(data[i])

print("Original list:", data)
print("Reversed list:", reversed_list)
```

```python
## 4...


values = [8, 2, 10, 5, 17, 3, 14]
maximum = values[0] 
for num in values:
    if num > maximum:
        maximum = num

print("Maximum element:", maximum)
```

```python
## 5....


items = ['red', 'blue', 'green', 'red', 'yellow', 'blue', 'red']
count = 0

for item in items:
    if item == 'red':
        
        count = count + 1

print("List:", items)
print("The string 'red' appears", count, "times in the list")
```

```python
## 7...



def check_palindrome(lst):
    if lst == lst[::-1]:
        print("Palindrome")
    else:
        print("Not a Palindrome")
        
check_palindrome([1, 2, 3, 2, 1])     
check_palindrome([1, 2, 3, 4, 5])     
check_palindrome(['a', 'b', 'a'])      
check_palindrome([])                  
```

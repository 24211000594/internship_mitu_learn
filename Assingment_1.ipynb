```python
# 1. Take two numbers as input from the user and print their sum, difference, product, and
#quotient — each on a separate line. Handle division by zero with an appropriate message

num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
print("Sum =", num1 + num2)
print("Difference =", num1 - num2)
print("Product =", num1 * num2)
# quotient
if num2 != 0:
    print("Quotient =", num1 / num2)
else:
    print("Division by zero is not possible")

```

```python
# 2.  Accept a temperature in Celsius from the user, convert it to Fahrenheit and Kelvin, and
# print all three values with labels.

celsius = float(input("Enter temperature in Celsius: "))
fahrenheit = (celsius * 9/5) + 32
kelvin = celsius + 273.15
print("Celsius =", celsius)

print("Fahrenheit =", fahrenheit)

print("Kelvin =", kelvin)

```

```python
# 5.Take a year as input and print whether it is a leap year or not.

year = int(input("Enter a year: "))

if year % 4 == 0:
    if year % 100 == 0:
        if year % 400 == 0:
            print(year, "is a leap year")
        else:
            print(year, "is not a leap year")
    else:
        print(year, "is a leap year")
else:
    print(year, "is not a leap year")
```

```python
# 6.Build a simple calculator — take two numbers and an operator (+, -, *, /) as input,
 #perform the operation, and print the result. Print an error message for invalid operators or
 #division by zero.

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

operator = input("Enter operator (+, -, *, /): ")

if operator == "+":
    print("Result =", num1 + num2)

elif operator == "-":
    print("Result =", num1 - num2)

elif operator == "*":
    print("Result =", num1 * num2)

elif operator == "/":
    if num2 != 0:
        print("Result =", num1 / num2)
    else:
        print("Error: Division by zero is not possible.")
else:
    print("Error: Invalid operator.")

    
```

```python
# 7. Take a number as input and print its multiplication table from 1 to 10. Format each line
 # as: 7 x 3 = 21

num = int(input("Enter a number: "))
for i in range(1, 11):
    print(num, "x", i, "=", num * i)
```

```python
# 8. . Take a positive integer n as input and print all prime numbers from 2 to n.

n = int(input("Enter a number: "))

for num in range(2, n + 1):
    is_prime = True

    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        print(num)
        
```

```python
# 9..  Keep accepting numbers from the user until they enter 0. Then print the total count of
 # numbers entered, their sum, and their average. Do not include 0 in the calculations.
count = 0
sum = 0
while True:
    num = int(input("Enter a number: "))

    if num == 0:
        break
    count = count + 1
    sum = sum + num

if count > 0:
    average = sum / count
else:
    average = 0

print("Total numbers entered:", count)
print("Sum:", sum)
print("Average:", average)
```

```python
# 10.Take n as input and print a triangle pattern of n rows made of asterisks, where row i
 # has (2i - 1) stars centered appropriately.


n = int(input("Enter number of rows: "))

for i in range(1, n + 1):
    
    stars = 2 * i - 1
    spaces = n - i
    print(" " * spaces, end="")
    print("*" * stars)



```

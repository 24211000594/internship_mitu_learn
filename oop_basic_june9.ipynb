```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        self.skills = []
    
    def add_skill(self, skill):
        self.skills.append(skill)
    
    def view_info(self):
        print(self.name, self.age, self.skills)

s1 = Student("Alice", 20)
s2 = Student("Bob", 22)
s3 = Student("Charlie", 21)

s1.add_skill("Python")
s2.add_skill("Java")
s3.add_skill("ML")

s1.view_info()
s2.view_info()
s3.view_info()
```

# Inheritance

```python
class Person:
    insti = "MITU"

    def __init__(self,age,ht):
        self.age = age
        self.ht = ht

    def show(self):
        print(f" age -> {self.age}, height -> {self.ht},institute -> {self.insti}")
```

```python
class ravi(Person):
    def __init__(self,age,ht,sal):
       super().__init__(age, ht)
       self.sal = sal

    def show_sal(self):
        print(f"salary is {self.sal}")
```

```python
a = ravi(25,5.9,57000)
```

```python
a.show_sal()
```

```python
a.show()
```

```python
class person:

    def __init__(self , age,name):

        self.__age = age
        self.__name = name

        # getter
    def get_name(self):
        return self.__name

    def get_age(self):
        return self.__age

        #setter
    def set_age(self,age):
        self.__age  = age

    def __str__(self):
        return f"age ->{self.__age},name-> {self.__name}"
```

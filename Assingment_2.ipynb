```python
# 1.. Write a function analyze_sentence(sentence) that takes a sentence and returns a dictionary with
# word_count, char_count (excluding spaces), unique_words (case-insensitive), and
# most_frequent_word. Test it with:
# "The quick brown fox jumps over the lazy dog the fox"


def analyze_sentence(sentence):
    words = sentence.lower().split()

    word_count = len(words)
    char_count = len(sentence.replace(" ", ""))
    unique_words = len(set(words))

    frequency = {}
    for word in words:
        frequency[word] = frequency.get(word, 0) + 1

    most_frequent_word = max(frequency, key=frequency.get)

    return {
        "word_count": word_count,
        "char_count": char_count,
        "unique_words": unique_words,
        "most_frequent_word": most_frequent_word
    }

sentence = "The quick brown fox jumps over the lazy dog the fox"
result = analyze_sentence(sentence)
print(result)

```

```python
# 2. Given the string below, do the following and print each result:
# text = "A man a plan a canal Panama"
# a) Reverse the string without using slicing.
# b) Check if it is a palindrome, ignoring case and spaces.
# c) Replace every vowel with *.
# d) Count how many times each vowel appears.


text = "A man a plan a canal Panama"

reversed_text = ""
for char in text:
    reversed_text = char + reversed_text
print("a) Reversed String:", reversed_text)

cleaned = text.replace(" ", "").lower()
reversed_cleaned = cleaned[::-1]
print("b) Is Palindrome  :", cleaned == reversed_cleaned)

vowels = "aeiouAEIOU"
starred = ""
for char in text:
    if char in vowels:
        starred += "*"
    else:
        starred += char
print("c) Vowels as *    :", starred)

print("d) Vowel Counts   :")
for v in "aeiou":
    count = text.lower().count(v)
    print(f"   {v} → {count}")


```

```python
# 3.. Write a function format_report(data) that takes a list of tuples (name, score, grade) and prints a
#formatted table where name is left-aligned (width 15), score is right-aligned (width 6), and grade is
#centered (width 8). Use this data:
#data = [("Alice", 92, "A"), ("Bob", 78, "B+"), ("Charlie", 55, "C"), ("Diana",
#100, "A+")].




def format_report(data):
    print("Name".ljust(15), "Score".rjust(6), "Grade".center(8))
    print("-" * 32)

    for name, score, grade in data:
        print(name.ljust(15), str(score).rjust(6), grade.center(8))

data = [
    ("Alice", 92, "A"),
    ("Bob", 78, "B+"),
    ("Charlie", 55, "C"),
    ("Diana", 100, "A+")
]

format_report(data)
```

```python
# 4. Write a function list_ops(nums) that:
# a) Removes duplicates while preserving the original order.
 # b) Returns a new list with only elements at even indices, squared.
# c) Flattens the nested list into a single list using list comprehension.
# nums = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5] nested = [[1, 2, 3], [4, 5], [6, 7, 8,
# 9]]



def list_ops(nums, nested):
    unique = []
    for i in nums:
        if i not in unique:
            unique.append(i)

    squared = []
    for i in range(len(nums)):
        if i % 2 == 0:
            squared.append(nums[i] * nums[i])
    flat = []
    for sublist in nested:
        for item in sublist:
            flat.append(item)

    print("Unique List:", unique)
    print("Squared Even Index Elements:", squared)
    print("Flattened List:", flat)
    
nums = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]
nested = [[1, 2, 3], [4, 5], [6, 7, 8, 9]]

list_ops(nums, nested)
```

```python
# 7. Write a function inventory_manager(records) that takes a list of tuples (product, category,
#quantity, price) and returns a dictionary grouped by category. Each category should have
#total_items, total_value, and a list of product names. Use:
#records = [ ("Laptop", "Electronics", 10, 55000), ("Mouse", "Electronics", 50,
#499), ("Desk", "Furniture", 8, 12000), ("Chair", "Furniture", 15, 4500),
#("Pen", "Stationery", 200, 20), ]



def inventory_manager(records):
    inventory = {}
    for product, category, quantity, price in records:
        inventory.setdefault(category, {"total_items": 0, "total_value": 0, "products": []})
        inventory[category]["total_items"] += quantity
        inventory[category]["total_value"] += quantity * price
        inventory[category]["products"].append(product)

    return inventory
records = [
    ("Laptop", "Electronics", 10, 55000),
    ("Mouse",  "Electronics", 50,   499),
    ("Desk",   "Furniture",    8, 12000),
    ("Chair",  "Furniture",   15,  4500),
    ("Pen",    "Stationery", 200,    20),
]

for category, data in inventory_manager(records).items():
    print(f"\nCategory     : {category}")
    print(f"  Products   : {data['products']}")
    print(f"  Total Items: {data['total_items']}")
    print(f"  Total Value: Rs.{data['total_value']}")
```

```python
# 8.8. Implement the following:
#a) A function power_all(*args, exp=2) that returns a list of each number raised to exp.
#b) A function profile(**kwargs) that prints each key-value pair as Key : Value.
#c) Use a lambda inside sorted() to sort the list of dicts below by cgpa in descending order.
#students = [{"name": "Arjun", "cgpa": 8.4}, {"name": "Sanya", "cgpa": 9.1},
#{"name": "Dev", "cgpa": 7.8}]

def power_all(*args, exp=2):
    result = []
    for num in args:
        result.append(num ** exp)
    return result

print("Power All:", power_all(2, 3, 4, 5, exp=3))

def profile(**kwargs):
    for key, value in kwargs.items():
        print(key, ":", value)

print("\nProfile Output:")
profile(name="Arjun", age=21, course="Data Science")

students = [
    {"name": "Arjun", "cgpa": 8.4},
    {"name": "Sanya", "cgpa": 9.1},
    {"name": "Dev", "cgpa": 7.8}
]
sorted_students = sorted(students, key=lambda x: x["cgpa"], reverse=True)
print("\nSorted Students:")
for s in sorted_students:
    print(s)





```

```python
#9. Create a module stats_utils.py with three functions:
#a) mean(data) — arithmetic mean.
#b) variance(data) — population variance without using the statistics module.
#c) normalize(data) — min-max normalization returning values between 0 and 1.
#In a separate main.py, import and use these functions on the dataset below. Also compute the
#standard deviation using the math module.
#dataset = [12, 45, 7, 23, 56, 89, 34, 67, 11, 90]

def mean(data):
    return sum(data) / len(data)


def variance(data):
    avg = mean(data)
    total = 0

    for x in data:
        total += (x - avg) ** 2

    return total / len(data)   # Population variance


def normalize(data):
    min_val = min(data)
    max_val = max(data)

    normalized = []

    for x in data:
        value = (x - min_val) / (max_val - min_val)
        normalized.append(value)

    return normalized
```

```python
import math

dataset = [12, 45, 7, 23, 56, 89, 34, 67, 11, 90]

avg = mean(dataset)
var = variance(dataset)
std_dev = math.sqrt(var)
norm_data = normalize(dataset)

print("Dataset:", dataset)
print("Mean:", avg)
print("Variance:", var)
print("Standard Deviation:", std_dev)
print("Normalized Data:", norm_data)
```

```python
# 10. Complete the following:
#a) Write a closure make_multiplier(n) that returns a function multiplying its input by n. Create triple
#and quintuple using it.
#b) Write a decorator @timer that prints the execution time of any function it wraps. Apply it to
#slow_sum(n) which computes the sum of squares from 1 to n.
#c) Write a decorator @validate_positive that raises a ValueError if any argument is zero or negative.
#Apply it to safe_sqrt(x).





import math
import time

# a) Closure
def make_multiplier(n):
    return lambda x: x * n

triple = make_multiplier(3)
quintuple = make_multiplier(5)

print(triple(10))
print(quintuple(10))


# b) Timer Decorator
def timer(func):
    def wrapper(n):
        start = time.time()
        result = func(n)
        print("Time:", time.time() - start)
        return result
    return wrapper

@timer
def slow_sum(n):
    return sum(i**2 for i in range(1, n+1))

print(slow_sum(10000))


# c) Validate Positive Decorator
def validate_positive(func):
    def wrapper(x):
        if x <= 0:
            raise ValueError("Number must be positive")
        return func(x)
    return wrapper

@validate_positive
def safe_sqrt(x):
    return math.sqrt(x)

print(safe_sqrt(25))

```

```python
#### 1..1. Create a class BankAccount with the following:
#a) Attributes: account_holder, account_number, and balance.
#b) Methods: deposit(amount), withdraw(amount), and get_balance().
#c) withdraw() should print a message and do nothing if the amount exceeds the current balance.
#d) A __str__ method that returns a readable summary of the account.
#Create two BankAccount objects and perform a few deposits and withdrawals on each


class BankAccount:
    def __init__(self, name, acc_no, balance):
        self.name = name
        self.acc_no = acc_no
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        if amount > self.balance:
            print("Not enough balance")
        else:
            self.balance -= amount

    def get_balance(self):
        return self.balance

    def __str__(self):
        return f"Name: {self.name}, Account No: {self.acc_no}, Balance: {self.balance}"

a1 = BankAccount("Rahul", 101, 5000)
a2 = BankAccount("Priya", 102, 3000)

# Transactions
a1.deposit(1000)
a1.withdraw(2000)

a2.deposit(500)
a2.withdraw(4000)

print(a1)
print(a2)
print(a2)
```

```python
## 2...Create a class Student with attributes name, roll_no, and marks (a list of 5 subject marks). Add
#methods:
#a) calculate_percentage() — returns the average of all marks.
#b) get_grade() — returns 'A' for >= 80, 'B' for >= 60, 'C' for >= 40, and 'Fail' otherwise.
#c) display() — prints all details including percentage and grade.
#Create a list of 3 Student objects and call display() on each

class Student:
    def __init__(self, name, roll_no, marks):
        self.name = name
        self.roll_no = roll_no
        self.marks = marks

    def calculate_percentage(self):
        return sum(self.marks) / 5

    def get_grade(self):
        p = self.calculate_percentage()
        if p >= 80:
            return "A"
        elif p >= 60:
            return "B"
        elif p >= 40:
            return "C"
        return "Fail"

    def display(self):
        print(self.name, self.roll_no, self.calculate_percentage(), self.get_grade())

students = [
    Student("Rahul", 1, [80, 85, 90, 75, 70]),
    Student("Priya", 2, [60, 65, 70, 55, 60]),
    Student("Amit", 3, [30, 40, 35, 45, 38])
]

for s in students:
    s.display()
```

```python
# 3...Section B — Encapsulation
#3. Create a class Employee with private attributes __name, __emp_id, and __salary. Add:
#a) Getter and setter methods for all three attributes.
#b) The salary setter should raise a ValueError if the value is below 10000.
#c) A method give_raise(percent) that increases salary by the given percentage.
#Demonstrate that the private attributes cannot be accessed directly, and that all access goes
#through the methods



class Employee:
    def __init__(self, name, emp_id, salary):
        self.__name = name
        self.__emp_id = emp_id
        self.set_salary(salary)

    def get_name(self): return self.__name
    def get_emp_id(self): return self.__emp_id
    def get_salary(self): return self.__salary

    def set_name(self, name): self.__name = name
    def set_emp_id(self, emp_id): self.__emp_id = emp_id

    def set_salary(self, salary):
        if salary < 10000:
            raise ValueError("Salary must be >= 10000")
        self.__salary = salary

    def give_raise(self, percent):
        self.__salary += self.__salary * percent / 100

e = Employee("Amit", 101, 20000)

print("Name:", e.get_name())
print("ID:", e.get_emp_id())
print("Salary:", e.get_salary())

e.set_salary(25000)
e.give_raise(10)

print("Updated Salary:", e.get_salary())

# Demonstration: private attribute access fails
print(e.__salary) 
```

```python
#4..4. Build a class SecureVault that stores a private __pin and a private __data dictionary. It should
have:
#a) A method add_entry(pin, key, value) that adds to __data only if the pin matches.
#b) A method get_entry(pin, key) that returns the value only if the pin matches.
#c) Any wrong pin attempt should print 'Access Denied' and return nothing



class SecureVault:
    def __init__(self, pin):
        self.__pin = pin
        self.__data = {}

    def add_entry(self, pin, key, value):
        if pin == self.__pin:
            self.__data[key] = value
        else:
            print("Access Denied")

    def get_entry(self, pin, key):
        if pin == self.__pin:
            return self.__data.get(key)
        else:
            print("Access Denied")

vault = SecureVault(1234)

vault.add_entry(1234, "name", "Amit")
vault.add_entry(1111, "city", "Pune")  

print(vault.get_entry(1234, "name"))
print(vault.get_entry(1111, "name"))   
```

```python
#5.5. Create a base class Shape with a method area() that returns 0 and a method describe() that
#prints 'I am a shape'. Create three subclasses:
#a) Circle(radius) — overrides area() using pi * r^2.
#b) Rectangle(length, width) — overrides area().
#c) Triangle(base, height) — overrides area().
#Each subclass should also override describe() to print something specific about itself. Create one
#object of each and call both methods...


class Shape:
    def area(self):
        return 0

    def describe(self):
        print("I am a shape")

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius

    def describe(self):
        print("I am a circle")

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def describe(self):
        print("I am a rectangle")

class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return 0.5 * self.base * self.height

    def describe(self):
        print("I am a triangle")

c = Circle(5)
r = Rectangle(4, 6)
t = Triangle(3, 8)

c.describe()
print("Circle area:", c.area())

r.describe()
print("Rectangle area:", r.area())

t.describe()
print("Triangle area:", t.area())
```

```python
## 6.








# Parent class
class Vehicle:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def start_engine(self):
        print("Engine started")


# Child class Car
class Car(Vehicle):
    def __init__(self, make, model, year, num_doors):
        super().__init__(make, model, year)   # Call parent constructor
        self.num_doors = num_doors
    def start_engine(self):
        print(f"{self.make} {self.model}: Engine started")

# Child class ElectricCar
class ElectricCar(Vehicle):
    def __init__(self, make, model, year, battery_capacity):
        super().__init__(make, model, year)   # Call parent constructor
        self.battery_capacity = battery_capacity

    def start_engine(self):
        print(f"Silent start - battery at {self.battery_capacity} kWh")

# Creating objects
v1 = Vehicle("Honda", "City", 2020)
c1 = Car("Toyota", "Corolla", 2022, 4)
e1 = ElectricCar("Tesla", "Model 3", 2024, 75)

v1.start_engine()
c1.start_engine()
e1.start_engine()
```

```python
#10



class StudentManager:
    def __init__(self):
        self.students = []

    def add(self):
        name = input("Name: ")
        roll = input("Roll No: ")
        m1 = int(input("Mark 1: "))
        m2 = int(input("Mark 2: "))
        m3 = int(input("Mark 3: "))
        self.students.append([name, roll, m1, m2, m3])

    def view(self):
        for s in self.students:
            avg = (s[2] + s[3] + s[4]) / 3
            print(s[0], s[1], "Average =", avg)

    def search(self):
        roll = input("Enter Roll No: ")
        for s in self.students:
            if s[1] == roll:
                print("Found:", s)
                return
        print("Not Found")

    def delete(self):
        roll = input("Enter Roll No: ")
        self.students = [s for s in self.students if s[1] != roll]
        print("Deleted")

manager = StudentManager()

while True:
    print("\n1.Add 2.View 3.Search 4.Delete 5.Exit")
    ch = input("Choice: ")

    if ch == "1":
        manager.add()
    elif ch == "2":
        manager.view()
    elif ch == "3":
        manager.search()
    elif ch == "4":
        manager.delete()
    elif ch == "5":
        break
    else:
        print("Invalid Choice")
```

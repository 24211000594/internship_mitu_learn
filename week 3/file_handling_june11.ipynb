```python
# write into a myfile.txt with context manager 
# use write function and add 3 separate lines
```

```python

with open("myfile.txt", "w") as f:
    f.write("My name is yash\n")
    f.write("I am learning Python\n")
    f.write("Python is easy and fun\n")
    
```

```python
with open("myfile.txt", "r") as f:
    data = f.read()
print(data)
```

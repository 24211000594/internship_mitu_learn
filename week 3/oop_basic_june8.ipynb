```python
class Employee:

    company = "ABC Pvt Ltd"

    def __init__(self, name, monthly_salary, job_designation):
        self.name = name
        self.__monthly_salary = monthly_salary  
        self.job_designation = job_designation
    def calculate_annual_salary(self):
        return self.__monthly_salary * 12
    def show_data(self):
        print("Company          :", Employee.company)
        print("Name             :", self.name)
        print("Designation      :", self.job_designation)
        print("Monthly Salary   :", self.__monthly_salary)
        print("Annual Salary    :", self.calculate_annual_salary())
        print("-" * 40)


# Creating 3 objects
e1 = Employee("Jatin", 50000, "Python Developer")
e2 = Employee("Sachin", 60000, "Data Analyst")
e3 = Employee("Rahul", 70000, "Project Manager")

e1.show_data()
e2.show_data()
e3.show_data()
```

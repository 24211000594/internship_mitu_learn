```python
# loan approval expert system 
```

```python
def approval_loan(income, credit_score,existing_loan):
    if income >= 50000 and credit_score >= 700 and not existing_loan:
        return "loan Approved"
    elif income >= 30000 and credit_score >= 650:
        return "further review required"
    else:
        return "loan rejected"
        
```

```python
income = 700000
credit_score = 750
existing_loan = False
```

```python
# Inference engine
```

```python
decision = approval_loan(
    income,
    credit_score,
    existing_loan
)
```

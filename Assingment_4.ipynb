```python
##1.




def safe_divide(a, b):
    try:
        print("Result:", a / b)
    except ZeroDivisionError:
        print("Cannot divide by zero.")
    except TypeError:
        print("Inputs must be numbers.")
    finally:
        print("Operation attempted.")

# Test Cases
safe_divide(10, 2)
safe_divide(10, 0)
safe_divide(10, "2")
```

```python
## 2
# Custom Exception Class
class InvalidAgeError(Exception):
    pass

# Function to validate age
def validate_age(age):
    if age < 0 or age > 120:
        raise InvalidAgeError("Invalid Age!")
    else:
        print("Valid Age:", age)

# Test with 3 different values
ages = [25, -5, 150]

for age in ages:
    try:
        validate_age(age)
    except InvalidAgeError as e:
        print("Error:", e)
```

```python
#3..
user_inputs = ['10', 'abc', '0', '5']
for value in user_inputs:
    try:
        number = int(value)      
        result = 100 / number    
        print("Result:", result)
    except ValueError:
        print("Error: Input is not a valid number.")
    except ZeroDivisionError:
        print("Error: Cannot divide by zero.")
    except Exception as e:
        print("Error:", e)
```

```python
##4...



import numpy as np
arr = np.arange(1, 21)
print("Original 1D Array:")
print(arr)
arr = arr.reshape(4, 5)
print("\n4x5 Array:")
print(arr)
# (b) Print the sum of each row and each column
print("\nSum of Each Row:")
print(np.sum(arr, axis=1))

print("\nSum of Each Column:")
print(np.sum(arr, axis=0))

# (c) Extract elements greater than 10
print("\nElements Greater Than 10:")
print(arr[arr > 10])

# (d) Replace all even numbers with -1
arr[arr % 2 == 0] = -1

print("\nArray After Replacing Even Numbers with -1:")
print(arr)
```

```python
###5..

import numpy as np

A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

# (a) Element-wise operations
print("Addition:")
print(np.add(A, B))

print("\nSubtraction:")
print(np.subtract(A, B))

print("\nMultiplication:")
print(np.multiply(A, B))

# (b) Matrix Multiplication
print("\nMatrix Multiplication:")
print(np.dot(A, B))

# (c) Transpose of A
print("\nTranspose of A:")
print(np.transpose(A))

# (d) Determinant of A
print("\nDeterminant of A:")
print(np.linalg.det(A))

# Inverse of A
print("\nInverse of A:")
print(np.linalg.inv(A))
```

```python
##6..

import numpy as np
np.random.seed(42)
arr = np.random.randint(1, 101, 15)

print("Original Array:")
print(arr)

# (a) Mean, Median, Standard Deviation, Minimum, Maximum
print("\nMean:", np.mean(arr))
print("Median:", np.median(arr))
print("Standard Deviation:", np.std(arr))
print("Minimum:", np.min(arr))
print("Maximum:", np.max(arr))

# (b) Sort the array in descending order
print("\nArray in Descending Order:")
print(np.sort(arr)[::-1])

# (c) Indices of the 3 largest values
print("\nIndices of 3 Largest Values:")
print(np.argsort(arr)[-3:])
normalized = (arr - np.min(arr)) / (np.max(arr) - np.min(arr))

print("\nNormalized Array:")
print(normalized)
```

```python
###7


import pandas as pd

# Create DataFrame
data = {
    "Name": ["Aarav", "Priya", "Karan", "Meera", "Sahil"],
    "Age": [23, 21, 25, 22, 24],
    "Department": ["AI", "Data Science", "AI", "Web Dev", "Data Science"],
    "Score": [88, 92, 79, 95, 67]
}

df = pd.DataFrame(data)
# (a) Display first 3 rows
print("First 3 Rows:")
print(df.head(3))
# Display data types
print("\nData Types:")
print(df.dtypes)
# (b) Add Grade column
grade = []
for score in df["Score"]:
    if score >= 90:
        grade.append("A")
    elif score >= 75:
        grade.append("B")
    else:
        grade.append("C")
df["Grade"] = grade
print("\nDataFrame with Grade:")
print(df)
# (c) Sort by Score in descending order
print("\nSorted by Score:")
print(df.sort_values(by="Score", ascending=False))

# (d) Filter rows where Department is AI
print("\nDepartment = AI:")
print(df[df["Department"] == "AI"])
```

```python
##8


import pandas as pd

data = {
    "Name": ["Aarav", "Priya", "Karan", "Meera", "Sahil"],
    "Age": [23, 21, 25, 22, 24],
    "Department": ["AI", "Data Science", "AI", "Web Dev", "Data Science"],
    "Score": [88, 92, 79, 95, 67]
}

df = pd.DataFrame(data)

# (a) Average score of each department
print(df.groupby("Department")["Score"].mean())

# (b) Row with highest score
print(df.loc[df["Score"].idxmax()])
# (c) Rename Score to Marks
df = df.rename(columns={"Score": "Marks"})
print(df)
# (d) Drop Age column
df = df.drop("Age", axis=1)
print(df)




```

```python
###9..

import pandas as pd
import numpy as np

data = {
    "Product": ["Pen", "Notebook", "Eraser", "Sharpener", "Ruler"],
    "Price": [10, np.nan, 5, 8, np.nan],
    "Stock": [100, 50, np.nan, 30, 20]
}

df = pd.DataFrame(data)

# (a) Check missing values
print(df.isnull().sum())
# (b) Fill missing Price with mean
df["Price"] = df["Price"].fillna(df["Price"].mean())
# (c) Drop rows where Stock is missing
df = df.dropna(subset=["Stock"])
# (d) Save to CSV
df.to_csv("cleaned_products.csv", index=False)

print(df)
```
